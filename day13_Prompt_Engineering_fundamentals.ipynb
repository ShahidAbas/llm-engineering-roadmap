# Day 13 Prompt Engineering Fundamentals

## What is Prompt Engineering?

Prompt Engineering is the process of designing clear and effective instructions (prompts) for Large Language Models (LLMs) to generate accurate and useful responses.

A well-written prompt improves the quality, accuracy, and format of the model's output.

Prompt = Instruction + Context + Input + Expected Output

```python
from google import genai
```

**Simple Prompt**

```python
client = genai.Client(api_key="YOUR_API_KEY")

response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents="What is Machine Learning?"
)

print(response.text)
```

**Better Prompt**

```python
response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents="""
Explain Machine Learning.

Rules:
- Maximum 3 lines.
- Use simple English.
- Give one real-world example.
"""
)

print(response.text)
```

**Zero-Shot Prompting**


The model receives only an instruction without any examples.

Example:

Translate:
Hello

```python
response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents="""
Classify the sentiment.

Text:
I love studying AI.
"""
)

print(response.text)
```

## One-Shot Prompting

The model receives one example before solving the task.

```python
response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents="""
Example:

Text:
I love this phone.

Sentiment:
Positive

Now classify:

Text:
I hate waiting in traffic.

Sentiment:
"""
)

print(response.text)
```

## Few-Shot Prompting

The model receives multiple examples before answering.

```python
response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents="""
Example 1

Text:
I love AI.

Sentiment:
Positive

Example 2

Text:
The weather is terrible.

Sentiment:
Negative

Example 3

Text:
Today is amazing.

Sentiment:
Positive

Now classify:

Text:
This movie is boring.

Sentiment:
"""
)

print(response.text)
```

## Role Prompting

Assigning a role helps the LLM tailor its response.

Example:

"You are a Python teacher."

"You are a recruiter."

"You are a research assistant."

```python
response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents="""
You are an AI professor.

Explain Deep Learning.
"""
)

print(response.text)
```

**Output Formatting**

```python
response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents="""
Explain RAG.

Answer only in bullet points.
"""
)

print(response.text)
```

**Markdown Table**

```python
response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents="""
Compare BERT and GPT.

Return the answer as a Markdown table.
"""
)

print(response.text)
```

**JSON Format**

```python
response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents="""
Explain Machine Learning.

Return the answer in JSON format.
"""
)

print(response.text)
```
