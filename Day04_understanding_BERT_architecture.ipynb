**Exercise 1:**

```python
from transformers import AutoModel, AutoTokenizer

model_name = "bert-base-uncased"  #Loads BERT's tokenizer (converts text → token IDs)

tokenizer = AutoTokenizer.from_pretrained(model_name)   #Loads the full BERT model with all its weights (~110M parameters)

model = AutoModel.from_pretrained(model_name)

print(model)
```

```python
print(model.config)
```

**Exercise 2**

```python
text = "i love machine learning"

inputs = tokenizer(text, return_tensors = "pt")

print(inputs)
```

```python
outputs = model(**inputs) #** unpacks a dictionary into keyword arguments

print(outputs.last_hidden_state.shape)
```

**Q1:** What problem does BERT solve compared to older NLP models?   
Answer:   
BERT understands the context and meaning of words by looking at surrounding words, whereas older NLP models often treated words independently and  
struggled with ambiguity.   

**Q2:** What is self-attention?    
Answer:    
Self-attention is a mechanism that allows a token to focus on and learn from other relevant tokens in the same sentence to better understand context   
and relationships.   

**Q3:** Why is BERT called bidirectional?    
Answer:    
BERT is called bidirectional because it reads text from both left-to-right and right-to-left simultaneously, allowing it to understand context from    both directions.  

**Q4:** What are transformer layers?    
Answer:     
Transformer layers are the building blocks of BERT that process embeddings using self-attention and neural networks to learn increasingly complex    
language patterns.   

**Q5:** What does torch.Size([1, 6, 768]) represent?   
Answer:   
1   → Batch size (1 sentence)   
6   → Number of tokens   
768 → Features learned for each token     

It means BERT produced a 768-dimensional representation for each of the 6 tokens in the sentence.
