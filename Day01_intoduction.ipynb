### NLTK   
stands for Natural Language Toolkit.   
It is a Python library used for working with human language data.
* __what it does:__ Provides easy-to-use functions for tokenization (splitting text), stemming, lemmatization, tagging parts of speech, and basic classification.
* __When to use:__ Great for learning NLP fundamentals, prototyping, and working with small datasets.
* __When NOT to use:__ Avoid for production or large-scale data—it is slow and outdated compared to modern libraries (like SpaCy or Hugging Face Transformers).

### TF-IDF and BERT  
* TF-IDF = Term Frequency – Inverse Document Frequency
* BERT = Bidirectional Encoder Representations from Transformers 

`TF-IDF` (Term Frequency - Inverse Document Frequency) is a numerical statistic used to evaluate how important a specific word is to a particular document within a larger collection of documents, known as a corpus. It works by combining two metrics: Term Frequency (TF), which counts how many times a word appears in a `single` document, and `Inverse` Document Frequency (IDF), which measures how rare or common that word is across the entire corpus. Words that appear frequently in one document but rarely in others receive a high TF-IDF score, effectively highlighting them as key terms that distinguish that document from the rest. Conversely, common stopwords like "the," "is," or "and" that appear everywhere are heavily penalized with low scores. As a result, TF-IDF is widely used in information retrieval, search engines, and text mining to extract meaningful keywords and rank document relevance. **However**, its major `limitation` is that it completely `ignores` word order, context, and semantic meaning, treating each word independently.

`BERT` (Bidirectional Encoder Representations from Transformers) is a deep learning model developed by Google for natural language processing that revolutionized how machines understand text. Unlike previous models that read words sequentially (left-to-right or right-to-left), BERT is bidirectional, meaning it reads the entire sentence in both directions simultaneously, allowing it to capture the full context of a word based on all surrounding words. It is pre-trained on massive amounts of text using a technique called Masked Language Modeling, where random words in a sentence are hidden, and the model learns to predict them based on their context. This enables BERT to generate contextual embeddings, meaning the same word gets a different vector representation depending on the sentence it appears in (e.g., "bank" in "river bank" vs. "money bank" have different embeddings). As a result, **BERT** achieves `state`-`of`-`the`-`art` performance on tasks like question answering, sentiment analysis, and named entity recognition. **However**, it is computationally `expensive`, requires significant `memory`, and is slower to train and deploy compared to simpler models like TF-IDF or Word2Vec.

#### Why TF-IDF has limitations?  
* **No Semantics:** It only detects word importance via frequency, not meaning. "Car" and "Automobile" are treated as completely different words.
* **Sparse Vectors:** It creates massive vectors with mostly zeros, which is memory-heavy and inefficient.
* **Out-of-Vocabulary (OOV):** It cannot handle words it has never seen during training.

#### What embeddings are?   
Embeddings are `dense`, `continuous` `vector` `representations` of words (or sentences). They map words into a `high-dimensional` space where semantic similarity = geometric distance (i.e., words with similar meanings are located close to each other).

## Word2Vec:   
`Word2Vec` is a shallow neural network model developed by Google that learns dense, fixed-length vector representations (embeddings) for words by analyzing their co-occurrence patterns with surrounding words in a large corpus. It operates on the principle that words appearing in similar contexts tend to have similar meanings, producing vectors where semantic and syntactic relationships are preserved (e.g., "king - man + woman ≈ queen"). However, Word2Vec assigns a single static vector to each word, meaning it cannot distinguish between different meanings of the same word based on context.

### Transformer:  
Transformer is a deep learning architecture introduced by Google in the 2017 paper "Attention Is All You Need" that completely revolutionized natural language processing. Unlike previous models like RNNs or LSTMs that processed text sequentially (word-by-word), the Transformer processes an entire sentence in parallel using a mechanism called self-attention. This self-attention mechanism allows the model to weigh the importance of every word in the input against every other word simultaneously, capturing long-range dependencies and contextual relationships regardless of their distance in the sentence. The architecture consists of an encoder (which reads the input) and a decoder (which generates output), both built from multiple layers of self-attention and feed-forward neural networks. Transformers are highly scalable, enabling training on massive datasets, and they form the backbone of virtually all modern state-of-the-art NLP models. BERT specifically uses only the encoder part of the Transformer, while models like GPT use only the decoder. **However**, Transformers are computationally `expensive`, requiring significant memory and specialized hardware like GPUs or TPUs, and they need enormous amounts of training data to perform effectively.

```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis")

result = classifier("I am excited to learn LLM Engineering.")

print(result)
```

#### Why is BERT better than TF-IDF?   
BERT is `superior` to `TF-IDF` because it fundamentally understands language rather than just counting words. TF-IDF treats text as a collection of independent keywords, completely ignoring word order, grammar, and context, meaning it cannot distinguish between sentences like "The dog chased the cat" and "The cat chased the dog" since both contain the same words. BERT, on the other hand, processes text bidirectionally using self-attention, capturing the full meaning of each word based on its surrounding context, allowing it to understand relationships, negation, sarcasm, and polysemy. For example, TF-IDF would assign the same score to "bank" in "river bank" and "money bank," while BERT generates completely different vectors for each based on the sentence. Additionally, BERT handles out-of-vocabulary words using subword tokenization, whereas TF-IDF simply drops unknown terms. This makes BERT dramatically more accurate on complex tasks like question answering, sentiment analysis, and named entity recognition. However, TF-IDF remains faster, lighter, and more interpretable, making it preferable for simple keyword-based search and applications with limited computational resources.

### Tokenization and its role:
`Tokenization` is the process of breaking down raw text into smaller, meaningful units called tokens. These tokens can be words, subwords, characters, or punctuation marks, depending on the level of granularity required. It is the very first and most crucial step in any NLP pipeline because machines cannot understand raw text; they can only process numbers. Tokenization converts text into a structured format that can then be mapped to numerical representations like embeddings or TF-IDF vectors for further processing.   
### The Role of Tokenization:   
* **For TF-IDF:** Tokenization splits text into individual words (or n-grams), which are then counted and weighted. Simple word-level tokenization works fine here because TF-IDF only cares about frequency, not meaning.
* **For Word2Vec:** Tokenization provides the sequence of words needed to train the model by analyzing co-occurrence patterns within a sliding window. Word-level tokenization is standard here.
* **For BERT and Transformers:** Tokenization plays a much more critical role. BERT uses subword tokenization (specifically WordPiece), which breaks rare or unknown words into smaller, frequent subword units (e.g., "unhappiness" → "un", "happiness" or "un", "happi", "ness"). This solves the out-of-vocabulary problem, allowing BERT to handle words it has never seen during training by splitting them into known pieces. Additionally, BERT tokenizers add special tokens like [CLS] (for classification) and [SEP] (to separate sentences), which are essential for the model to understand sentence boundaries and perform downstream tasks correctly.

```python
### testing work
from transformers import pipeline

classifier = pipeline("sentiment-analysis")

texts = [
    "I love machine learning",
    "I hate waiting in traffic",
    "Today is an ordinary day"
]

for text in texts:
    print(text)
    print(classifier(text))
    print("-" * 50)
```

### Tokenization:   
A token is the smallest meaningful unit of text that an NLP model processes. It is the output produced by a tokenizer after breaking down raw text. Depending on the tokenization strategy, a token can be:
* **A word:** (e.g., "apple", "running", "NLP")
* **A subword:** (e.g., "un", "happi", "ness" from "unhappiness")
* **A character:** (e.g., "a", "b", "c")
* **A punctuation mark:** (e.g., ".", "!", "?")
For example, the sentence "I love NLP!" tokenized at the word level becomes: ["I", "love", "NLP", "!"] — resulting in 4 tokens.

In the context of BERT and Transformer models, tokens are not just raw text fragments; they are immediately mapped to unique numerical IDs (token IDs) that correspond to entries in the model's vocabulary. These IDs are then fed into the model to lookup their corresponding embeddings. Subword tokens like "un" and "happi" are especially valuable because they allow the model to understand and process rare or completely new words by breaking them down into familiar pieces, effectively solving the out-of-vocabulary problem. In simple terms, tokens are the basic building blocks that machines use to read and understand human language.

### Why do LLMs use tokens instead of words?  
**LLMs** use tokens instead of whole words primarily to balance vocabulary size, computational efficiency, and handling of unseen words. If an LLM used whole words, it would need a vocabulary of hundreds of thousands or even millions of words to cover all variations, including plurals, verb tenses, misspellings, and rare domain-specific terms. This massive vocabulary would make the model extremely large, slow, and memory-intensive, with most words appearing very rarely, wasting capacity.  
**Subword** tokenization (used by models like BERT and GPT) solves this by breaking words into smaller, frequently occurring pieces. For example, instead of storing "running," "runner," and "runs" as three separate words, the model stores the root "run" and suffixes like "ning," "ner," and "s," combining them as needed. This drastically reduces the vocabulary size (typically 30,000–100,000 tokens) while still covering virtually any text.   
**Additionally**, subword tokens handle out-of-vocabulary words gracefully. If the model encounters a word it has never seen, like "tokenizationist," it can break it down into known subwords like "token," "ization," and "ist" instead of throwing an error. This is especially crucial for handling misspellings, scientific jargon, and new slang. Finally, tokens also serve as a compression mechanism, as common words like "the" become single tokens while rare words may split into multiple, making the overall sequence length manageable for the model's fixed context window.
