An embedding is a dense numerical vector that captures the meaning of a word or sentence.

```python
from sentence_transformers import SentenceTransformer
```

```python
model = SentenceTransformer("all-MiniLM-L6-v2")
print("Loading model...")

embedding = model.encode("I learn Machine Learning")
print("Model loaded!")

print(type(embedding))
print(embedding.shape)
```

**This means the sentence became a vector containing 384 numbers.**

#### Exercise 2 compare similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
s1 = model.encode('I learn AI')
s2 = model.encode('I enjoiy Artificial Intelligence')
s3 = model.encode('I hat traffic jams')

print('s1 vs s2:', 
     cosine_similarity([s1],[s2])[0][0])
print('s1 vs s3:', 
     cosine_similarity([s1],[s3])[0][0])
```

### Exercise 3 

```python
sentences = ["I love coding",
             "Conding is my passion",
             "The weather is normal today",
             "Python is amazing language"
            ]
# generate embedding for all sentences at once
embeddings = model.encode(sentences)

print('embedding shape: ', embeddings.shape)

# compare every sentence with every other sentence
print("\nPairwise similarity matrix:")
similarity_matrix = cosine_similarity(embeddings)

for i in range(len(sentences)):   #use loop for all sentences
    for j in range(len(sentences)):
        if i < j :            # avoid repeating pairs and self-comparison
            print(f"'{sentences[i]}' vs '{sentences[j]}': {similarity_matrix[i][j]:.4f}")
```

### Exercise 4

```python
import numpy as np

documents = [
    "I love programming",
    "Coding is my passion",
    "The weather is hot today",
    "Python is an amazing language",
    "I enjoy hiking in the mountains",
    "Machine learning models need data"
]

doc_embeddings = model.encode(documents)

# A search query
query = "I really like writing code"
query_embedding = model.encode(query)

# Compare query to all documents
similarities = cosine_similarity([query_embedding], doc_embeddings)[0]

# Find the best match
best_idx = np.argmax(similarities)

print(f"Query: {query}\n")
for doc, score in zip(documents, similarities):
    print(f"{score:.4f}  →  {doc}")

print(f"\nBest match: '{documents[best_idx]}' (score: {similarities[best_idx]:.4f})")
```

```python
documents = [
    "I love programming",
    "Coding is my passion",
    "The weather is hot today",
    "Python is an amazing language",
    "I enjoy hiking in the mountains",
    "Machine learning models need data"
]

doc_embeddings = model.encode(documents)

query = "I really like writing code"
query_embedding = model.encode(query)

similarities = cosine_similarity([query_embedding], doc_embeddings)[0]

# --- Top 3 matches ---
top_n = 3
top_indices = np.argsort(similarities)[::-1][:top_n]

print(f"Query: {query}\n")
print(f"Top {top_n} matches:")
for rank, idx in enumerate(top_indices, start=1):
    print(f"{rank}. {documents[idx]}  (score: {similarities[idx]:.4f})")
```
