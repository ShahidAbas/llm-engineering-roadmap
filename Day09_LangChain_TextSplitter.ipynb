#### What is a Text Splitter?   

Think of it as:    

Large Document   
        ↓   
LangChain Text Splitter    
        ↓   
Chunk 1  
Chunk 2   
Chunk 3   
Chunk 4   
... and so on   

It automatically divides large documents into manageable chunks.  

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
text = """
Artificial Intelligence is a branch of computer science.

Machine Learning is a subset of AI.

Deep Learning is a subset of Machine Learning.

Natural Language Processing enables computers to understand human language.

Computer Vision enables machines to understand images.

Python is the most popular programming language for AI.
"""
```

**Create a text Spliter**

```python
splitter = RecursiveCharacterTextSplitter(
    chunk_size=100,
    chunk_overlap=20
)
```

**spliting text**

```python
chunks = splitter.split_text(text)

print(chunks)
```

**print chunks**

```python
for i, chunk in enumerate(chunks):
    print(f"Chunk {i+1}")
    print(chunk)
    print("-"*50)
```
