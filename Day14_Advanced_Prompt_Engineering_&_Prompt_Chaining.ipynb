# Day 14 Advanced Prompt Engineering

## What is Advanced Prompting?

Advanced Prompting is the process of designing prompts that guide Large Language Models (LLMs) to produce more accurate, structured, and task-specific responses.

Instead of asking simple questions, we provide:
- Clear instructions
- Context
- Constraints
- Output format
- Examples

This significantly improves the quality of the generated response.

**Poor Prompt**

```python
from google import genai
```

```python
client = genai.Client(api_key="Your_API_KEY")
response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents="Explain AI."
)

print(response.text)
```

**Better Prompt**

```python
response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents="""
You are an AI instructor.

Explain Artificial Intelligence.

Rules:
- Maximum 50 words
- Beginner-friendly
- Include one real-life example
- Use bullet points
"""
)

print(response.text)
```

## Prompt Template   
- Instead of writing prompts repeatedly, we create reusable templates.

```python
topic = "Machine Learning"

prompt = f"""
You are an AI teacher.

Explain {topic}.

Rules:
- Beginner-friendly
- Maximum 5 bullet points
- Include one practical example.
"""

response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents=prompt
)

print(response.text)
```

**Step-by-Step Prompting**  
Large problems become easier if we ask the model to solve them one step at a time.

```python
prompt = """
Explain how an email spam detector works.

Answer using these steps:

1. Collect Data
2. Clean Data
3. Train Model
4. Predict
5. Improve Model
"""

response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents=prompt
)

print(response.text)
```

**Prompt Chaining**   
Prompt Chaining means using the output of one prompt as the input to another.  
Example:  

Prompt 1  
   ↓  
Summary  
   ↓  
Prompt 2  
   ↓  
Quiz   
   ↓   
Prompt 3  
   ↓   
Evaluation  
This is common in AI workflows.

**Step 1**

```python
response1 = client.models.generate_content(
    model="gemini-2.5-flash",
    contents="Explain Machine Learning in one paragraph."
)

summary = response1.text

print(summary)
```

**Step 2**

```python
response2 = client.models.generate_content(
    model="gemini-2.5-flash",
    contents=f"""
Using the following explanation:

{summary}

Create five interview questions.
"""
)

print(response2.text)
```

**Step 3**

```python
response3 = client.models.generate_content(
    model="gemini-2.5-flash",
    contents=f"""
Using these interview questions:

{response2.text}

Provide short model answers.
"""
)

print(response3.text)
```

**this how LLM work flow is working**

# Output Constraints
LLMs perform better when you specify the expected output.

#### Bullet form

```python
prompt = """
Explain RAG.

Return only bullet points.
"""
```

#### Table form

```python
prompt = """
Compare BERT and GPT.

Return the answer as a Markdown table.
"""
```

#### JASON

```python
prompt = """
Explain FAISS.

Return the response in JSON format.
"""
```

## Error Correction

```python
code = """
def add(a,b)
return a+b
"""

prompt = f"""
Find the errors in the following Python code and fix them.

{code}
"""

response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents=prompt
)

print(response.text)
```

- This demonstrates using LLMs as coding assistants.
