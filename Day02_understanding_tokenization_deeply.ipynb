#### Why Can't Models Understand Text?   
Suppose we give BERT:  

**I love Machine Learning**
  
Humans understand it immediately.   

Computers only understand:   

0, 1, 2, 3, ...  

So we need a `translator`:  

Text  
 ↓  
Tokenizer  
 ↓   
Numbers   
 ↓    
Transformer Model   

The `tokenizer` is that `translator`.

### Let's Inspect BERT's Tokenizer

```python
from transformers import AutoTokenizer
```

```python
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

text = "I am learning LLM engineering."

tokens = tokenizer.tokenize(text)
print(tokens)
```

**Notice:**   
ll   
##m   

BERT didn't know "LLM" as a complete word.   

So it split it into smaller pieces.    

`This is called:`   
Subword Tokenization

### Why Subword Tokenization?   
Imagine a vocabulary containing:   
cat  
dog   
house   
book  
`Now we encounter:`   
ChatGPT   
What should the model do?   
`Instead of failing:`   
ChatGPT   
`becomes`   
Chat   
##GP   
##T   
or similar.   
`This allows models to understand:`   
New words   
Misspellings   
Rare words   
`without retraining.`

### Convert Text to IDs

```python
encoded = tokenizer("I am learning LLM engineering")
print(encoded)
```

`input_ids`   
These are token numbers.    
as we have 'input_ids': [101, 1045, 2572, 4083, 2222, 2213, 3330, 102]   
`Think:`    
Word → Token → ID   
`Example:`    
learning → 4083  

### Decode IDs Back

```python
print(encoded["input_ids"])
```

```python
decoded = tokenizer.decode(encoded["input_ids"])
```

```python
print(decoded)
```

What is tokenization?    
Tokenization is the process of breaking text into smaller units called tokens so that a language model can process and understand the text.

Why do LLMs use tokens instead of words?   
LLMs use tokens because tokens can represent complete words, parts of words, punctuation, and special symbols. This helps the model handle unknown words, rare words, and multiple languages more efficiently.

What are special tokens?  
Special tokens are predefined tokens with specific purposes, such as marking the start, end, or padding of a sequence. Examples include [CLS], [SEP], and [PAD].

What are input_ids?   
input_ids are the numerical representations of tokens. Since neural networks work with numbers, each token is converted into a unique ID before being fed into the model.

What is attention_mask?   
attention_mask tells the model which tokens should be considered (1) and which tokens should be ignored (0), especially when padding is used.
