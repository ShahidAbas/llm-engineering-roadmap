#### Semantic Search with FAISS  

FAISS stand for `Facebook AI Similarity Search`

Today we will build a mini search engine that searches by meaning, not keywords.

```python
from sentence_transformers import SentenceTransformer
```

```python
#It converts any sentence into a 384-dimensional vector that captures its meaning.
model = SentenceTransformer("all-MiniLM-L6-v2")

sen_doc = [
    "I love Artificial Intelligence",
    "Machine Learning is fascinating",
    "Today is very hot",
    "Deep Learning is a branch of AI",
    "Python is a programming language"
]

#Passes all sentences through the model. Each sentence becomes a vector of 384 numbers. Similar sen == simi vectors.
embeddings = model.encode(sen_doc)

print(embeddings.shape)
```

5  sentences    
384 feature search or dimensions

#### Build FAISS Index or Vector Database

```python
import faiss
import numpy as np

dimension = embeddings.shape[1]

index = faiss.IndexFlatL2(dimension)

index.add(np.array(embeddings))
```

What happened?   
Documents   
 ↓   
Embeddings   
 ↓   
Stored in FAISS  

We just created a vector database.

#### Search query in the Database

```python
#create query
query = "I enjoy AI"

query_embedding = model.encode([query])

#search query
distances, indices = index.search(
    np.array(query_embedding),
    k=2
)

print(indices)
```

```python
#retrieve results
for idx in indices[0]:
    print(sen_doc[idx])
```

```python
sen_doc = [
    "I want to play footbal",
    "I love Artificial Intelligence",
    "Machine Learning is fascinating",
    "Today is very hot",
    "AI",
    "Deep Learning is a branch of AI",
    "Python is a programming language",
    "Artificial Intelligence",
    "Programming",
    "Weather",
    "Neural Networks"
]


embeddings = model.encode(sen_doc)

print('sentences and dimensions:',embeddings.shape)

dimension = embeddings.shape[1]

index = faiss.IndexFlatL2(dimension)

index.add(np.array(embeddings))

#create query
query = "I enjoy AI"

query_embedding = model.encode([query])

#search query
distances, indices = index.search(
    np.array(query_embedding),
    k=10
)

print('indexes:',indices)

#retrieve results
for idx in indices[0]:
    print(sen_doc[idx])
```
